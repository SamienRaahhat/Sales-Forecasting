### Adding Additional Features

```python
# Holiday flag: 1 if type_x contains 'Holiday', else 0

df_merged['is_holiday'] = df_merged['type_x'].apply(lambda x: 1 if 'Holiday' in str(x) else 0)
df_merged
```

```python
# Government payday flag: 1 if 15th or last day of the month
df_merged['is_govt_payday'] = df_merged['day'].apply(lambda x: 1 if x == 15 else 0)

# Identify last day of the month dynamically
df_merged['last_day_of_month'] = df_merged['date'] + pd.offsets.MonthEnd(0)

# Add flag for last day of the month
df_merged['is_last_day'] = (df_merged['date'] == df_merged['last_day_of_month']).astype(int)

# Final payday flag: 1 if either 15th or last day of the month
df_merged['is_govt_payday'] = df_merged[['is_govt_payday', 'is_last_day']].max(axis=1)
```

### Rolling Statistics

```python
# Ensure 'date' column is sorted
df_merged = df_merged.sort_values(by=['store_nbr', 'date'])

# Compute 7-day moving average
df_merged['sales_movingavg_7'] = df_merged.groupby('store_nbr')['sales'].transform(lambda x: x.rolling(window=7, min_periods=1).mean())

# Compute 30-day moving average
df_merged['sales_movingavg_30'] = df_merged.groupby('store_nbr')['sales'].transform(lambda x: x.rolling(window=30, min_periods=1).mean())
```

```python
# Compute rolling standard deviation for past 7 days

df_merged['sales_std_7'] = df_merged.groupby('store_nbr')['sales'].transform(lambda x: x.rolling(window=7, min_periods=1).std())
```

```python
# Previous week's sales (lag of 7 days)
df_merged['sales_lag_7'] = df_merged.groupby('store_nbr')['sales'].shift(7)

# Previous month's sales (lag of 30 days)
df_merged['sales_lag_30'] = df_merged.groupby('store_nbr')['sales'].shift(30)
```

```python
# Compute average sales per store type
avg_sales_per_store_type = df_merged.groupby('store_nbr')['sales'].mean()

print(avg_sales_per_store_type)
```

```python
# Compute total sales per family and cluster
topselling_productfamilies_per_cluster = df_merged.groupby(['cluster', 'family'])['sales'].sum().reset_index()

# Find the top product family per cluster
topselling_productfamilies_per_cluster = topselling_productfamilies_per_cluster.sort_values(['cluster', 'sales'], ascending=[True, False]).groupby('cluster').first()

print(topselling_productfamilies_per_cluster)
```

### 1) ARIMA

```python
# Aggregate sales per date (since ARIMA works on a single time series)

train_df_arima = train_df.groupby('date')['sales'].sum()
test_df_arima = test_df.groupby('date')['sales'].sum()
```

```python
from pmdarima import auto_arima

# Auto-select best ARIMA(p, d, q) parameters
arima_model = auto_arima(train_df_arima, seasonal=False, trace=True, suppress_warnings=True)

# Fit final ARIMA model
arima_model.fit(train_df_arima)
```

```python
# Forecasting sales for the next 15 days
future_forecast = arima_model.predict(n_periods=len(test_df_arima))

# Convert predictions to DataFrame
forecast_df = pd.DataFrame({'date': test_df_arima.index, 'predicted_sales': future_forecast})

# Merge with actual sales for comparison
result_df = test_df_arima.reset_index().merge(forecast_df, on='date', how='left')
```

```python
result_df.head(3)
```

### Model Evaluation

```python
# Evaluating ARIMA Performance

from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score

# Extract actual and predicted values
y_actual = result_df['sales']
y_pred_arima = result_df['predicted_sales']

# Compute Evaluation Metrics
rmse_arima = np.sqrt(mean_squared_error(y_actual, y_pred_arima))
mape_arima = mean_absolute_percentage_error(y_actual, y_pred_arima)
r2_arima = r2_score(y_actual, y_pred_arima)

print(f"ARIMA Performance:\nRMSE: {rmse_arima}\nMAPE: {mape_arima}\nR² Score: {r2_arima}")
```

### ARIMA Visual Inspection - Actual Sales Vs Predicted Sales

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(result_df['date'], result_df['sales'], label="Actual Sales", color='blue')
plt.plot(result_df['date'], result_df['predicted_sales'], label="Predicted Sales (ARIMA)", color='red')
plt.xticks(rotation=45)
plt.legend()
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Actual vs Predicted Sales - ARIMA Model")
plt.show()
```

### Takeaways
- Since the **R²** is **-0.48**, the ARIMA model is likely not capturing trends and seasonality properly.
- **High RMSE & High MAPE** means that the Model is not accurate enough for business use.
- **Negative R²** means that the The model is performing poorly, worse than a simple average prediction.
- Hence, **ARIMA** may not be the best choice for this dataset.

### 2) XGBoost

```python
train_df['day'] = train_df['date'].dt.day
train_df['month'] = train_df['date'].dt.month
train_df['year'] = train_df['date'].dt.year
train_df['dayofweek'] = train_df['date'].dt.dayofweek

test_df['day'] = test_df['date'].dt.day
test_df['month'] = test_df['date'].dt.month
test_df['year'] = test_df['date'].dt.year
test_df['dayofweek'] = test_df['date'].dt.dayofweek

X_train = train_df.drop(columns=['sales', 'date'])  # Keep the new numerical columns
y_train = train_df['sales']

X_test = test_df.drop(columns=['sales', 'date'])
y_test = test_df['sales']
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score

# Initialize XGBoost Model
xgb_model = xgb.XGBRegressor(objective="reg:squarederror", n_estimators=500, learning_rate=0.05, max_depth=6)

# Train the model
xgb_model.fit(X_train, y_train)

# Make Predictions
y_pred = xgb_model.predict(X_test) 

```

### Model Evaluation

```python
# Caluclating metrics

from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Making Predictions
y_pred = xgb_model.predict(X_test)

# Computing RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Fixing MAPE issue (avoiding division by zero)
threshold = 70  # Ignore sales <= 70 units (tuning as needed)
non_zero_indices = y_test > threshold

if np.any(non_zero_indices):
    mape = np.mean(np.abs((y_test[non_zero_indices] - y_pred[non_zero_indices]) / y_test[non_zero_indices])) * 100
else:
    mape = np.nan  # Avoid division issues

# Computing R² Score
r2 = r2_score(y_test, y_pred)

# Printing Metrics
print(f"XGBoost Performance:")
print(f"RMSE: {rmse}")
print(f"MAPE: {mape}")
print(f"R² Score: {r2}")
```

### Actual Sales vs Predicted Sales

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(y_test.values, label="Actual Sales", color="blue")
plt.plot(y_pred, label="Predicted Sales", color="red")
plt.xlabel("Time")
plt.ylabel("Sales")
plt.title("Actual vs. Predicted Sales - XGBoost")
plt.legend()
plt.show()
```

### Takeaways:
- **High R² (0.837)** suggests strong trend capture.
- **Reasonable RMSE** for a large dataset.
- **High MAPE (55.24%)** indicates poor accuracy in lower sales ranges.
- Some predictions might be overestimating or underestimating sales.

### 3) LSTM

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))  # Scale between 0 and 1
```

```python
X_train_scaled = scaler.fit_transform(X_train)  # Fit on training data
X_test_scaled = scaler.transform(X_test)  # Transform test data
```

```python
y_train_scaled = scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = scaler.transform(y_test.values.reshape(-1, 1))
```

```python
# Reshape input features into [samples, time steps, features]

X_train_reshaped = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_reshaped = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import Huber

# Initialize the LSTM model
model = Sequential([
    LSTM(units=128, return_sequences=True, input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2])),
    Dropout(0.2),
    LSTM(units=64, return_sequences=False),
    Dropout(0.2),
    Dense(units=32, activation = 'relu'),
    Dense(units=1)  # Output layer for regression
])

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss=Huber())
```

```python
# Train the model
history = model.fit(X_train_reshaped, y_train_scaled, epochs=50, batch_size=32, validation_data=(X_test_reshaped, y_test_scaled), verbose=1)
```

```python
# Predict on test data
y_pred_scaled = model.predict(X_test_reshaped)

# Inverse transform predictions and actual values
y_pred = scaler.inverse_transform(y_pred_scaled)
y_test_actual = scaler.inverse_transform(y_test_scaled.reshape(-1, 1))  
```

### Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score

# RMSE
rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred))

# MAPE
non_zero_indices = y_test_actual.flatten() != 0
mape = np.mean(np.abs((y_test_actual.flatten()[non_zero_indices] - y_pred.flatten()[non_zero_indices]) / y_test_actual.flatten()[non_zero_indices])) * 100

# R² Score
r2 = r2_score(y_test_actual, y_pred)

print("LSTM Performance:")
print("RMSE:", rmse)
print("MAPE:", mape)
print("R² Score:", r2)
```

### Visual for Actual Sales vs Predicted Sales

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(y_test_actual, label="Actual Sales")
plt.plot(y_pred, label="Predicted Sales", linestyle="dashed")
plt.legend()
plt.title("Actual vs Predicted Sales - LSTM")
plt.show()
```

### Takeaways:
- R2 score of 0.77 shows its ability to capture trends
- High MAPE suggests poor generalization.
